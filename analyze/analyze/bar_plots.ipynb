```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

## Load data

```python
long_df = pd.read_csv('optimization-experiments.csv')
```

## Bar plots ALG x Reranker

### NDCG Loan bucket 5

```python
select_df = long_df[(long_df['Protected'] == 'Loan Bucket 5') & (long_df['Metric']=='NDCG') &
                    (long_df['Optimization'].isin(['Separate', 'Joint']))]
```

```python
sb.set(font_scale=1.75)
g = sb.catplot(x="Reranker", y='Value', col="Base", col_wrap=2, hue='Optimization',

                data=select_df,

                kind="bar", orient='v')

g.set_axis_labels("", "NDCG")
plt.savefig('loan_size_bar_ndcg.png', facecolor='w', edgecolor='w')
```

### PSP Loan Bucket 5

```python
select_df = long_df[(long_df['Protected'] == 'Loan Bucket 5') & (long_df['Metric']=='PSP') &
                    (long_df['Optimization'].isin(['Separate', 'Joint']))]
```

```python
g = sb.catplot(x="Reranker", y='Value', col="Base", col_wrap=2, hue='Optimization',

                data=select_df,

                kind="bar", orient='v')

g.set_axis_labels("", "PSP")
plt.savefig('loan_size_bar_psp.png', facecolor='w', edgecolor='w')
```
